```python
#導入模組
from sklearn.cluster import KMeans #K-means
import matplotlib.pyplot as plt
import numpy as np
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
```

```python
est = KMeans()
est.fit(X)
```

```python
labels = est.labels_
```

```python
ax1=X[:,0]
ax2=X[:,1]
ax3=X[:,2]
ax4=X[:,3]
```

```python
def get3Dplot(axis1,axis2,axis3,labels,Title=''):
    fig = plt.figure(2, figsize=(5, 4))
    ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=48, azim=134)
    ax.scatter(axis1,axis2,axis3,
               c=labels, edgecolor='k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(Title)
    plt.show()
```

```python
#比較不同的clusters
est=KMeans(n_clusters=3)
est.fit(X)
get3Dplot(ax3,ax4,ax2,est.labels_.astype(np.float),'clusters=3')
est=KMeans(n_clusters=6)
est.fit(X)
get3Dplot(ax3,ax4,ax2,est.labels_.astype(np.float),'clusters=6')
```

```python
#比較不同的init
est=KMeans(n_clusters=3,n_init=10)
est.fit(X)
get3Dplot(ax3,ax4,ax2,est.labels_.astype(np.float),'init10')
est=KMeans(n_clusters=3,n_init=10,init='random')
est.fit(X)
get3Dplot(ax3,ax4,ax2,est.labels_.astype(np.float),'init=random')
```

```python
# 導入真正的y，看資料分布狀況！
get3Dplot(ax3,ax4,ax2,y,'True Value')
```
