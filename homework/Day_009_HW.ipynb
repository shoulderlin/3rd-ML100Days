### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
def ecdf(data):
    x = np.sort(data)
#     print(x)
    y = np.arange(1, len(x)+1)/len(x)
    return (x,y)

def plot_ecdf(data):
    print(data.name)
    print(data.describe())
    x,y = ecdf(data)
    tcdf = np.sort(np.random.normal(data.mean(), data.std(), size=len(data)))
#     print(tcdf)
    plt.plot(x,y)
    plt.plot(tcdf,y)
    plt.xlabel('Value')
    plt.ylabel('ECDF')
#     plt.xlim([x.min(), x.max()])
    plt.show()
    plt.plot(np.log(x),y)
    plt.plot(np.log(tcdf),y)
#     plt.xlim([np.log(x.min()), np.log(x.max())])
    plt.show()
```

```python
# 最大值離平均與中位數很遠

# 繪製 Empirical Cumulative Density Plot (ECDF)


"""
YOUR CODE HERE
"""
plot_ecdf(app_train['AMT_INCOME_TOTAL'])

```
